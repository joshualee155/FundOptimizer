```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
target_rets = [0.03, 0.04, 0.05, 0.06, 0.07, 0.08, 0.09, 0.10]
```

```python
dfs = {target_ret: pd.read_csv(f'backtest_2020-01-01_2021-09-30_120_20_{target_ret}.csv', index_col=0, parse_dates=True) for target_ret in target_rets}
```

```python
for df in dfs.values():
    df.fillna(0.0).plot(legend=False)
```

```python
for tr, df in dfs.items():
    mean_ret = df.sum(axis=1).pct_change().mean() * 252
    std_ret = df.sum(axis=1).pct_change().std() * 252**0.5
    sharpe = mean_ret / std_ret
    df.sum(axis=1).plot(label=f"TR:{tr:.2f},Sharpe:{sharpe:.2f}")
    plt.legend()
    
```
