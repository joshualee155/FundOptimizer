```python
C = df[cols].cov()*12
M = df[cols].mean()*12**0.5
```

```python
import cvxpy as cvx
```

```python
x = cvx.Variable(shape=(5, ))

variance = cvx.quad_form(x, C)
returns = cvx.sum(x*M)

objective = cvx.Minimize(variance)

constraints = [cvx.sum(x) == 1.,
               x>= 0.0,
               returns >= 0.04]

prob = cvx.Problem(objective, constraints)
```

```python
prob.solve()
```

```python
def MaximizeSharpeRatio(df):
    C = df.cov()*12
    M = df.mean()*12**0.5
    
    N = len(M)
    
    x = cvx.Variable(shape=(N, ))
    variance = cvx.quad_form(x, C)
    returns = cvx.sum(x*M)

    objective = cvx.Minimize(variance)

    constraints = [x>= 0.0,
                   returns == 1.0]

    prob = cvx.Problem(objective, constraints)
    prob.solve()
    
    x_opt = x.value/x.value.sum()
    
    return x_opt
```

```python
cols = np.random.choice(df.columns, 50, replace=False)
x=MaximizeSharpeRatio(df[cols])
```

```python
port_return = (df[cols]*x).sum(axis=1)
cum_return = (port_return/20.+1.).cumprod()
px.line( port_return )
```
